```python
# Data loading
file_path2 = 'Resources/shopping_data.csv'
df_shopping= pd.read_csv(file_path2, encoding='ISO-8859-1')
df_shopping.head(5)
```

```python
# Columns
df_shopping.columns
```

```python
# List dataFrame data types
df_shopping.dtypes
```

```python
# Find null values
for column in df_shopping.columns:
    print(f'Column{column} has {df_shopping[column].isnull().sum()} null values' )
```

```python
# Drop null rows
df_shopping = df_shopping.dropna()
```

```python
# Find duplicate entries
print(f'Duplicate entries: {df_shopping.duplicated().sum()}')
```

```python
# Remove the customerID column
df_shopping.drop(columns=['CustomerID'], inplace=True)
df_shopping.head()
```

```python
# Transform string column
def change_string(member):
    if member == 'Yes':
        return 1
    else:
        return 0

df_shopping['Card Member']= df_shopping['Card Member'].apply(change_string)
df_shopping.head()
```

```python
# Transform annual Income
df_shopping['Annual Income']= df_shopping['Annual Income'] / 1000
df_shopping.head()
```

```python
new_df_shopping= df_shopping.rename(columns= {
    'Card Member' : 'card_member',
    'Annual Income' : 'annual_income',
    'Age': 'age',
    'Spending Score (1-100)' : 'spending_score'
})
new_df_shopping.head()
```

```python
# Saving cleaned data
file_path = "Resources/shopping_data_cleaned.csv"
new_df_shopping.to_csv(file_path, index=False)
```
